```python
#Import all the packages and ensure that pandas print all the rows

import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
df = pd.read_csv('polytechnics_data.csv') 
df
```

```python
#to find missing values in each column
df.isna().sum()
```

```python
#For all the empty cells in the dataframe, replace with 0

df[['year','sex','course','intake','enrolment','graduates']] = \
    df[['year','sex','course','intake','enrolment','graduates']].replace(0,np.NaN)
df
```

```python
# drop the Nan values
df.dropna()
df
```

```python
df[['year','sex','course','intake','enrolment','graduates']].to_csv('polytechnics_datacleaned1.csv', index=False)
```
